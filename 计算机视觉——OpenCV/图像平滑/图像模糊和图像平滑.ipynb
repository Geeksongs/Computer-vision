```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread("456.jpg")
show(image)
```

```python
kernelsizes=[(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,15))
for i,kenel in enumerate(kernelsizes):
    plt.subplot(1,3,i+1)
    #
    blur=cv2.blur(image,kenel)
    plt.axis('off')#
    plt.title('great title'+str(kenel))
    plt.imshow(blur)
plt.show()
```

```python
kernelsizes=[(3,3),(9,9),(15,15)]#
plt.figure(figsize=(15,15))
for i,kenel in enumerate(kernelsizes):
    plt.subplot(1,3,i+1)
    #
    blur=cv2.GaussianBlur(image,kenel,0)
    plt.axis('off')#
    plt.title('great title'+str(kenel))
    plt.imshow(blur)
plt.show()
```

```python
plt.figure(figsize=(15,15))
for i,kenel in enumerate((3,9,15)):#3*39*915*15
    plt.subplot(1,3,i+1)
    #
    blur=cv2.medianBlur(image,kenel,0)
    plt.axis('off')#
    plt.title('great title'+str(kenel))
    plt.imshow(blur)
plt.show()
```
