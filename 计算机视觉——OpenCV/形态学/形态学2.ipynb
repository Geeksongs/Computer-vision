```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
#下面开始opening开运算,腐蚀之后膨胀，用于去除白点
image=imread("123.jpg")
show(image)
```

```python
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(10,10))
opening=cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel)
show(opening)
```

```python
#下面开始闭运算,膨胀之后再腐蚀，用于去除黑点
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(10,10))
opening=cv2.morphologyEx(image,cv2.MORPH_CLOSE,kernel)
show(opening)
```

结果十分显然，把眼睛都去掉了

```python
#形态学梯度，用于得到图像的轮廓
opening=cv2.morphologyEx(image,cv2.MORPH_GRADIENT,kernel)
show(opening)
```

```python
#白帽运算，得到可以被去除的白色的点
image=imread('123.jpg')
opening=cv2.morphologyEx(image,cv2.MORPH_TOPHAT,kernel)
show(opening)
```

```python
#黑帽运算,得到可以被去除黑色的点
image=imread('123.jpg')
opening=cv2.morphologyEx(image,cv2.MORPH_BLACKHAT,kernel)
show(opening)
```
