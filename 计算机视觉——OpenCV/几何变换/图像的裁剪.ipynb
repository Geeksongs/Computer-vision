```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread("qwe.jpg")
show(image)#在定义了函数之后，图片的颜色终于正确了
```

```python
image=cv2.imread("qwe.jpg")
image=image[0:50,0:50]
```

```python
show(image)
```

```python
image=imread("qwe.jpg")
image=image[0:1792,0:450]
show(image)
```
