```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread('qwe.jpg')
image.shape#
```

上面的3表示了这个图片具有三个通道，分别是R,G,B

```python
(R,G,B)=cv2.split(image)#将图像进行切分
```

```python
print(R.shape)
print(G.shape)
print(B.shape)
```

```python
#现在开始图像的合并
merge=cv2.merge([R,G,B])
show(merge)
```
