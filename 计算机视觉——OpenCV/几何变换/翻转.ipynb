```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
image=cv2.imread("qwe.jpg")
show(image)
```

```python
image=cv2.flip(image,1)#这里用到的是水平翻转，因为后面的参数是一
show(image)
```

```python
image=cv2.flip(image,0)#这里用到的是垂直翻转，因为后面的参数是0
show(image)
```

```python
image=cv2.flip(image,0)#这里用到的是水平垂直翻转，因为后面的参数是-1
show(image)
```
