```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
image=cv2.imread("qwe.jpg")
show(image)
```

```python
image=cv2.flip(image,1)#
show(image)
```

```python
image=cv2.flip(image,0)#0
show(image)
```

```python
image=cv2.flip(image,0)#-1
show(image)
```
