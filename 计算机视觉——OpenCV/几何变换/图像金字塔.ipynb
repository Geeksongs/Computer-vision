```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
image=imread('qwe.jpg')
```

```python
for i in range(4):
    image=cv2.pyrDown(image)#这里做了一次高斯金字塔的计算
    print(image.shape)
    show(image)
```
